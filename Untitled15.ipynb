```python
import pandas as pd

# Load the dataset
file_path = '/train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Get basic information about the dataset
data.info()

# Check for missing values
data.isnull().sum()
```

```python
# Display the columns with missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]

# Fill missing values or drop rows/columns
# Example: Filling missing values in 'Age' with the median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Drop columns with too many missing values or not useful for analysis
data.drop(columns=['Cabin'], inplace=True)
```

```python
# Summary statistics for numerical variables
data.describe()

# Summary statistics for categorical variables
data.describe(include=['O'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Bar plot of Survived
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()

# Box plot of Age by Survived
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Age Distribution by Survival')
plt.show()
```
